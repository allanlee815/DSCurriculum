### Adding Columns

Adding columns is a very simple process! 

There are a few things you need in order to add a column:
1. You need a unique column name! 
    - This name will be used for the column's label 
2. You need a list of values to add to the column
    - There should be one value per row
    - Meaning the number of values = number of rows in our dataframe!
    
Then you simply use variable assignment to add the column:

```DataFrame['new_column_name'] = [list_of_values_for_new_column]```

*(You can also set all the rows of a column to a specific value. For example, `DataFrame['new_column_name'] = True`.)*

If you notice, the kill/death or K/D ratio is not part of our dataframe, so let's add it:

#### What is a K/D?
K/D ratio stands for the number of kills you make per the number of deaths you suffer. To find a K/D ratio, you simply take the number of kills and divide it by the number of deaths. 

For example, a K/D of 1 means you get one kill per death. K/D of 2 means you get 2 kills per death. 

**Can you see if we have the information for kills and deaths given in the table?**

````python
league.head()
````

### ```'blueDeaths'``` can be used for Deaths!

### ```'blueKills'``` can be used for Kills! 

All we have to do is divide the kills by the corresponding deaths! (That sounds gruesome. . .)

### Looking back to NumPy:

If you remember from NumPy, there is element-wise math happening. 

We can take advantage of this element by element computation to help us find the K/D ratio!

````python
kills = league['blueKills'].values
deaths = league['blueDeaths'].values

kd_ratio = kills/deaths
kd_ratio
````

*Note that this code cell gives a division by zero warning. That's because some games had zero blue deaths. So, it returns inf as the K/D. Please don't go around dividing by zero though.*

### The K/D for the first row is 1.5

The first element in our ```kd_ratio``` array is 1.5, which corresponds to the K/D ratio for the first row in our dataframe. 

Now to add this to our table, we use the tools from above: 

### We will call this column ```'K/D'```:

````python
# Creating the new column and assigning the corresponding values we computed earlier.

league['K/D'] = kd_ratio
league.head()
````

### Our new column is all the way to the right!

### Here is one-line of code that will produce the same result:

````python
league['New K/D'] = league['blueKills'] / league['blueDeaths']
````

````python
league.head()
````

### Applying Functions to Columns:

Another way to create a new column is by applying a function to a separate column in the DataFrame. The result will be the values for our new column. 

**Let's say we want there to be a column that reads ```'Win'``` if there was a win and ```'Loss'``` if there was a loss for that row.**

There is no function that will do that for us... so we will need to define our own. 

#### Can you find out which column(s) have information about wins and losses? 

````python
league.head()
````

### The second column, ```'blueWins'``` tells us if there was a win or a loss for each row. 

A win is shown as ```1```, and a loss is shown as ```0```. 

We will use this information to help us write a new column whose values are either ```'win'``` or ```'loss'```, depending on if that row experienced a win or loss. This will be done by applying a function to the ```"blueWins"``` column. 

### Our function works as follows....

If our function receives ```1``` as an argument, we will return ```'win'```.

If our function receives ```0``` as an argument, we will return ```'loss'```. 

### First lets try using a for loop###

````python
# This is not very clean and a bit chaotic. 

wins = []
for number in league['blueWins']:
    if number == 0:
        wins = wins + ['loss']
    elif number == 1:
        wins = wins + ['win']
league['win/loss?'] = wins
league.head()
````

````python
# We will now create a function to do this for us.
# So if we wanted to do something similar in the future,
# we can use this function instead of having to copy/past/edit 
# the code from above.

def win_loss(number):
    if number == 0:
        return 'loss'
    elif number == 1:
        return 'win'
````

````python
# Let's check if our function works as expected:

win_loss(0)
````

````python
win_loss(1)
````

Great, now that our function works, we can apply it to the ```blueWins``` column using ```.apply()```:

````python
league['blueWins'].apply(win_loss)

# This ends up being a lot cleaner
````

### This returned a list of ```'win'``` and ```'loss'``` values. 

Let's save that to a new column:

````python
league['win/loss?'] = league['blueWins'].apply(win_loss)
````

````python
league.head()
````

All the way to the right is our new ```win/loss?``` column!

### Lambda Functions ###

A lambda function is a small one time function. A lambda function can take any number of arguments, but can only have one expression. The difference is that they are not reusable like regular functions, but they can be done in one line.  

To create a lambda function, you must use the 'lambda' keyword followed by your arguments then a ':' to signal the start of your expression. Put together it should look like the following: 

```lambda arguments: expression```
    
To call your lambda function you can wrap it in parenthesis and then immediately after wrap your argument in parenthesis. 

For instance, ```(lambda arguments: expression)(argument)```

Note that the reason this is a 'one time' function is that there is no way to refer or reference your lambda function in any other code cell or line for that matter. 

````python
(lambda x:x*x)(10)
````

### Now lets put it all together ###

The lambda function can work with the ```.apply``` method as well. 

The general syntax is 

```table_name.apply(lambda arguments: expression), axis = 1) ```


### Next ###
We will use ```.apply``` method and the ```lambda``` function together to check whether the team that got the first blood won or not.  

````python
# First we make a normal function to see if they are the same or different

def BloodWin (blood, win):
    if (blood == win):
        return 'same'
    else:
        return 'different'
````

````python
# Now lets call it in a .apply method utilizing a lambda function

league['BloodWin'] = league.apply(lambda x: BloodWin(x['blueFirstBlood'], x['blueWins']), axis =1)
league.head(15)
````

## Recap

#### Sorting 
- If you would like to order the rows in your DataFrame based on the values of a certain column, you can call ```.sort_values("column_name")``` on your DataFrame, which will cause your DataFrame to reorder itself so the values in the certain column are listed in ascending order
    - If you want the values in the certain column to be listed in descending order, you have to call ```.sort_values("column_name", ascending = False)```
    
#### Adding Columns
- If you want to add a new column to your DataFrame, follow the format ```df_name["new_col_name"] = [list_of_values]```
- You can create a new column of values based off of the values in other columns thanks to element by element computation. Say you wanted a new column that had the sums of the values in two other columns, then you'd write ```df_name["new_col_name"] = df_name["other_col_name"] + df_name["other_other_col_name"]```
- You can also create a new column by applying a function to a separate column in the DataFrame to get the values for the new column, just follow the format ```df_name["new_col_name"] = league["separate_col_name"].apply(some_function)```

#### Lambda  functions
- Lambda functions take some input(s) and return an output, here is the syntax to create a lambda function:
  ```(lambda argument(s): expression)(argument(s))```
- The expressions inside of lambda functions can be almost anything! They can use other functions, dataframes, or even new lambda functions.  
- Lambda functions can also work with the ```.apply``` method to use multiple parameters in the format
  ```df_name.apply(lambda x: some_function(x[col_name1]...x[col_name for last parameter], axis=1)```
