### Dropping Columns
Often in our dataframe, we will not need some columns for our data analysis, so we will get rid of or **drop** them. 

Can you *identify* a column in our table that seems extra and particularly useless? 

````python
music.head()
````

Looks like the column **Unnamed: 0** seems to be useless. So let's drop it. 

Now here's where you can use the interent if you don't remember how to drop a column. Seach soemthing like ```"how to drop a column in pandas"```. 

This what I was linked to: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.drop.html. 

Here's what I took away: ```DataFrame.drop(labels=   , axis=   )```

Here, I need to replace ```DataFrame``` with the name of my frame, which is ```music```.

I need to fill in ```labels``` and ```axis``` to use this properly. 

In the documentation, it says to type 1 or type in ```'columns'``` for the ```axis```.

In ```labels``` we need to specify the label of the column:

````python
music.drop(labels='Unnamed: 0', axis=1)
````

````python
music.drop(labels='Unnamed: 0', axis='columns')
````

### Does not save it in place:
So these changes were not "saved". 

If I call my original dataframe music, the column is still there:

````python
music.head()
````

To make your dataframe **save** your column drop, you can do any of the following: 

### Use Inplace:
You can pass in another argument to the function (i.e. ```inplace=True```), which will do the operation "inplace" and save the change to your data table.

For example: ```music.drop(labels='Unnamed: 0', axis=1, inplace=True)```

### New Variable Assignment:
Instead of using the inplace argument, we can also simply assign ```music``` to be equal to the new dataframe that is returned after calling ```.drop```:

For example: ```music = music.drop(labels='Unnamed: 0', axis=1)```

*I prefer the second option, so that's what I usually do when coding.*

*Note*: I don't need to write ```labels``` in ```.drop``` because the computer will recognize that for me. Writing ```axis``` however is ***necessary***:

````python
music = music.drop('Unnamed: 0', axis=1)
````

````python
# Let's check to see if the column is still there? 

music.head()
````

### Null Values:

**Null values** are a common problem in Data Science. They're pretty easy to deal with, but can give you some trouble if you are not careful. 

<br>
<img src='../pimages/hardtimes.jpg' width="400"/>
<br>

**Fixing null values** is part of the process of Data Cleaning. 

### What is a Null Value?
A Null Value is represented as ```NaN``` which is used to indicate that there is a missing data value. There is no value that exists in our dataset. 

To check if there is a null value, you can use ```.isna()``` on your dataframe. What that does is that it asks each data value ``` is it null?```

````python
music.isna()
````

### No Null Values

Fortunately, we are all good and there are no null values in this dataset. ```False``` to ```.isna()``` means there is no null value there!

*Tip for doing data projects*: **Usually there are no null values if the data comes directly from a documentation or Kaggle.** 

### How to fix null values? 

Generally, the key way to fix null values is by filling it with ```0``` or some other specified value. 

That can be done with ```.fillna()``` which fills the null value!

We don't have any null values, but let's just to it anyways: 

````python
music.fillna(0) #it is always better to be safe than sorry
````

If you want to get rid of rows or columns with null values, you would use ```dropna(axis = )```. <br>

Like ```drop()```, you have to specify whether to drop the row (axis = 0) or column (axis = 1) that contains the null value.  

To make it pernament, you would add ```inplace = True``` or assign the DataFrame to the modified version too.

````python
music = music.dropna(axis = 'rows')
music
````

### More Manipulation:

Another data manipulation technique is to group together your rows by a certain attribute: 

For the ```music``` feature it makes sense to group the rows by artists and see the data for each ```artist``` in our frame. This will be useful in comparing artists and their music, potentially seeing who's the best! 

<br>
<img src='../pimages/dua.jpeg' width="500"/>
<br>

You can search ```how to group dataframe in pandas``` on the internet to find out how to do this! Try using the documentation! 

We will use ```.groupby()```. 

Here is the full documentation: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.groupby.html

````python
# We will groupby 'artists':

music.groupby('artist')
````

### Groupby Object

So pretty much "nothing" was returned after grouping... and that is normal! The weird looking result from the above cell is called a **groupby object**! 

This happened because we did not specify how we want to aggregate the other information. Sure, we are grouping by the artist...but what about the other columns? How should we group those for each artist? Should we just add them, average them, take the median, etc. 

For example, we can take the groupby object and get the average of the other columns. 

````python
music.groupby('artist').mean()
````

````python
# Unlocking to display the artists' averages/means in each column

music.groupby('artist').mean().loc['Drake']
````

````python
import numpy as np
min(music[music['artist'] == 'Drake']['acousticness'])
````

````python
# Unlocking to display the artists' sums in each column

music.groupby('artist').sum().loc['Drake']
````

You can also use `.agg()` for more.

````python
music.groupby('artist').agg('min').loc['Drake']
````

You can do more than one function with `agg()` too.

````python
energies = music.groupby('artist').energy.agg(['min', 'max'])
energies
````

Note that it just returns the same for both min and max if there's only one song for that artist (eg. !!!)

There's also another way to do this that comes in handy if you want *different functions for different columns.*

````python
# i.e. passing in a dictionary mapping specific columns to specific functions.

energies_custom = music.groupby('artist').agg({'loudness': 'min', 'speechiness': 'max'})
energies_custom
````

This is a dataframe that is now easier to work with because the index contains names of the artists. *We can use ```.loc``` for this.*

````python
energies.loc['Grimes']
````

````python
energies_custom.loc['Grimes']
````

Now let's get back to ```'Rihanna'``` and ```'Katy Perry'``` and compare. 

````python
grouped_music = music.groupby('artist').mean()
grouped_music.loc[['Rihanna', 'Katy Perry']]
````

# Recap

### Dropping data
- ```.drop(label = , axis = )``` drops a named row or column in the DataFrame. Type 1 or 'columns' in the axis to properly drop a column. To make the change pernament to the DataFrame, add ```inplace = True``` or assign the DataFrame to its modified counterpart.
- ```.dropna(axis = )``` drops rows or columns with missing values. Like ```drop()```, you need to specify the axis and make changes to save the data drop.
- ```.isna()``` checks the DataFrame for null values or NaN. You can fill these null values with ```n``` using ```filla(n)```

### Grouping data
- ```.groupby(named_column)``` groups data with the same value in a named column together. You can unlock this groupby object to display different values like the average (```.mean()```) or sum (```.sum()```).
- ```.agg()``` allows the user to perform different actions on different columns of a groupby object.
    <Br>ex: you can see the minimum loudness AND maximum speechiness of an artist.
- ```.loc[]``` displays the data associated with a specified label.
